```
# create based pyspark dataframe for data analysis
baseDF = sqlContext.sql("SELECT Year, id, price, host_is_superhost, neighbourhood_cleansed, instant_bookable, host_total_listings_count, latitude, longitude, property_type, room_type, accommodates, bathrooms, bedrooms, beds, minimum_nights, maximum_nights,  number_of_reviews,review_scores_rating, review_scores_accuracy, review_scores_cleanliness, review_scores_checkin, review_scores_communication, review_scores_location, review_scores_value, scrape_id FROM airbnb_listings")
display(baseDF), type(baseDF)
```

### 1) Data Cleansing

```
#Fixing PRICE Data Type
#price shows as string and for regression/ML analysis we need it to be a numeric = double type

from pyspark.sql.functions import col, translate

fixedPriceDF = baseDF.withColumn("price", translate(col("price"), "$,", "").cast("double"))

display(fixedPriceDF)
```

```
# Display Summary statistics
#summary (describe + IQR)
display(fixedPriceDF.summary())
```

```
# Display statistics using describe()
display(fixedPriceDF.summary())
```

```
# get rid of nulls in the categorical feature host_is_superhost
noNullsDF = fixedPriceDF.na.drop(subset=["host_is_superhost"])
```

```
#integer fields to double fixes - needed for ML
from pyspark.sql.functions import col
from pyspark.sql.types import IntegerType

integerColumns = [x.name for x in baseDF.schema.fields if x.dataType == IntegerType()]
doublesDF = noNullsDF

for c in integerColumns:
  doublesDF = doublesDF.withColumn(c, col(c).cast("double"))

columns = "\n - ".join(integerColumns)
print(f"Columns converted from Integer to Double:\n - {columns}")
```

```
# Create dummy variable if we will impute any value.
from pyspark.sql.functions import when

imputeCols = [
  "bedrooms",
  "bathrooms",
  "beds", 
  "review_scores_rating",
  "review_scores_accuracy",
  "review_scores_cleanliness",
  "review_scores_checkin",
  "review_scores_communication",
  "review_scores_location",
  "review_scores_value"
]

for c in imputeCols:
  doublesDF = doublesDF.withColumn(c + "_na", when(col(c).isNull(), 1.0).otherwise(0.0))
```

```
display(doublesDF.describe())
```

```
#inpute part - using median value
from pyspark.ml.feature import Imputer

imputer = Imputer(strategy="median", inputCols=imputeCols, outputCols=imputeCols)

imputedDF = imputer.fit(doublesDF).transform(doublesDF)
```

```
#check on price column
display(imputedDF.select("price").describe())
```

```
#check listings we can find where the price is zero
imputedDF.filter(col("price") == 0).count()
```

```
#filter out prices <0
posPricesDF = imputedDF.filter(col("price") > 0)
```

```
#check on min and max values of the minimum_nights column
display(posPricesDF
  .groupBy("minimum_nights").count()
  .orderBy(col("count").desc(), col("minimum_nights"))
)
```

```
# filter out values higher than one year
cleanDF = posPricesDF.filter(col("minimum_nights") <= 365)

display(cleanDF)
```

```
# trying to preserve the dataframe by saving it to a file so that we can start building models with it (just in case I need to split DBricks workbooks)

outputPath = "/tmp/DS3_G2_GA/airbnb-clean.parquet"

cleanDF.write.mode("overwrite").parquet(outputPath)
```

## 2) Regression: Predicting Rental Price

```
#Train/Test Split Phase -> 80% for the training set +  20% of our data for the test set.
trainDF, testDF = cleanDF.randomSplit([.8, .2], seed=42)
print(f"There are {trainDF.cache().count()} rows in the training set, and {testDF.cache().count()} in the test set")
```

```
# just checking on repartition our data to simulate how it could have been partitioned differently on a different cluster configuration, and see if we get the same number of data points in our training set.
(trainRepartitionDF, testRepartitionDF) = (cleanDF
                                           .repartition(24)
                                           .randomSplit([.8, .2], seed=42))

print(trainRepartitionDF.count())
```

```
#display trainDF
display(trainDF.select("price", "bedrooms").summary())
```

- it looks there are going to be some outliers present which might affect the model results and probalby will need to be taken care of later

```
#Vector Assembler - NOTE(JP): input column for Linear Regression expects a Vector type

from pyspark.ml.feature import VectorAssembler

vecAssembler = VectorAssembler(inputCols=["bedrooms"], outputCol="features")

vecTrainDF = vecAssembler.transform(trainDF)

vecTrainDF.select("bedrooms", "features", "price").show(10)
```

```
# Using Linear Regression estimator to build our first model
from pyspark.ml.regression import LinearRegression

lr = LinearRegression(featuresCol="features", labelCol="price")
lrModel = lr.fit(vecTrainDF)
```

```
# checking on the model 
m = round(lrModel.coefficients[0], 2)
b = round(lrModel.intercept, 2)

print(f" Linear regression line  formula: price = {m}*bedrooms + {b}")
```

```
# Using SPARK Pipeline
from pyspark.ml import Pipeline

pipeline = Pipeline(stages=[vecAssembler, lr])
pipelineModel = pipeline.fit(trainDF)

# applying to the test
predDF = pipelineModel.transform(testDF)

predDF.select("bedrooms", "features", "price", "prediction").show(10)
```

```
# Find average price on the training dataset ->   use the average price as our prediction column for our test dataset -> evaluate findings

from pyspark.sql.functions import avg, lit

trainDF, testDF = cleanDF.randomSplit([.8, .2], seed=42)

avgPrice = float(trainDF.select(avg("price")).first()[0])
predDF = testDF.withColumn("avgPrediction", lit(avgPrice))
```

```
from pyspark.ml.evaluation import RegressionEvaluator

regressionMeanEvaluator = RegressionEvaluator(predictionCol="avgPrediction", labelCol="price", metricName="rmse")

print(f"The RMSE for predicting the average price is: {regressionMeanEvaluator.evaluate(predDF):.2f}")
```

```
#adding additional features to our model using One-Hot Encoding for categorical variables
# use a combinantion of StringIndexer, OneHotEncoder, and VectorAssembler.

from pyspark.ml.feature import OneHotEncoder, StringIndexer

categoricalCols = [field for (field, dataType) in trainDF.dtypes 
                   if dataType == "string"]
indexOutputCols = [x + "Index" for x in categoricalCols]
oheOutputCols = [x + "OHE" for x in categoricalCols]

stringIndexer = StringIndexer(inputCols=categoricalCols, 
                              outputCols=indexOutputCols, 
                              handleInvalid="skip")
oheEncoder = OneHotEncoder(inputCols=indexOutputCols, 
                           outputCols=oheOutputCols)
```

```
# Combine OHE categorical features with  numeric features
from pyspark.ml.feature import VectorAssembler

numericCols = [field for (field, dataType) in trainDF.dtypes 
               if ((dataType == "double") & (field != "price"))]
assemblerInputs = oheOutputCols + numericCols
vecAssembler = VectorAssembler(inputCols=assemblerInputs, 
                               outputCol="features")
```

```
# build a linear regression model
from pyspark.ml.regression import LinearRegression

lr = LinearRegression(labelCol="price", featuresCol="features")
```

```
# put everything in a Pipeline
# Option 1: StringIndexer + OHE + VectorAssembler
from pyspark.ml import Pipeline

stages = [stringIndexer, oheEncoder, vecAssembler, lr]
pipeline = Pipeline(stages=stages)

pipelineModel = pipeline.fit(trainDF)
predDF = pipelineModel.transform(testDF)
predDF.select("features", "price", "prediction").show(5)
```

```
# Evaluate Model: RMSE eval
from pyspark.ml.evaluation import RegressionEvaluator

regressionEvaluator = RegressionEvaluator(predictionCol="prediction", labelCol="price", metricName="rmse")

rmse = round(regressionEvaluator.evaluate(predDF), 2)
print(f"RMSE is {rmse}")

#check on R2
r2 = round(regressionEvaluator.setMetricName("r2").evaluate(predDF), 2)
print(f"R2 is {r2}")
```

- Logarithmic scale

```
#Improving model performance by transforming the label to log scale ->  predicting on log scale -> evaluate the result.

from pyspark.sql.functions import col, log
from pyspark.ml import Pipeline
from pyspark.ml.feature import RFormula
from pyspark.ml.regression import LinearRegression

(trainDF, testDF) = cleanDF.randomSplit([.8, .2], seed=42)
logTrainDF = trainDF.withColumn("log_price", log(col("price")))
logTestDF = testDF.withColumn("log_price", log(col("price")))

rFormula = RFormula(formula="log_price ~ . - price", featuresCol="features", labelCol="log_price", handleInvalid="skip") 

lr = LinearRegression(labelCol="log_price", predictionCol="log_pred")
pipeline = Pipeline(stages = [rFormula, lr])
pipelineModel = pipeline.fit(logTrainDF)
predDF = pipelineModel.transform(logTestDF)
```

```
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.sql.functions import col, exp

expDF = predDF.withColumn("prediction", exp(col("log_pred")))

regressionEvaluator = RegressionEvaluator(labelCol="price", predictionCol="prediction")
rmse = regressionEvaluator.setMetricName("rmse").evaluate(expDF)
r2 = regressionEvaluator.setMetricName("r2").evaluate(expDF)
print(f"RMSE is {rmse}")
print(f"R2 is {r2}")
```
